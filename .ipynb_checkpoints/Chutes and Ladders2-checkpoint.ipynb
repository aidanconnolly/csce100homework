```python
import random

winner = False

players = {"player1": 0, "player2": 0, "player3": 0, "player4": 0}

def spinner():
    return random.randint(1,6)

def move(location):
    spin = spinner()
    playermove = location + spin

    if playermove == 1:
        playermove = 38
    elif playermove == 4:
        playermove = 14
    elif playermove == 9:
        playermove = 31
    elif playermove == 16:
        playermove = 6
    elif playermove == 21:
        playermove = 42
    elif playermove == 28:
        playermove = 84
    elif playermove == 36:
        playermove = 44
    elif playermove == 48:
        playermove = 26
    elif playermove == 49:
        playermove = 11
    elif playermove == 51:
        playermove = 67
    elif playermove == 56:
        playermove = 53
    elif playermove == 62:
        playermove = 19
    elif playermove == 65:
        playermove = 60
    elif playermove == 71:
        playermove = 91
    elif playermove == 80:
        playermove = 100
    elif playermove == 87:
        playermove = 24
    elif playermove == 93:
        playermove = 73
    elif playermove == 95:
        playermove = 75
    elif playermove == 98:
        playermove = 78
    elif playermove > 100:
        playermove = playermove - spin
        
    return playermove

while winner == False:
    for player, location in players.items():
        newLocation = move(location)
        players[player] = newLocation
        print(player, "moved to", players[player])
        if newLocation == 100:
            winner = True
        break
    else:
        continue
```
